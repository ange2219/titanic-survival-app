# 2.Préparation des données


## Netoyage

### Gestion des valeurs manqantes

```python
df.isna().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df = df.dropna()
df.isna().sum()
```

### Gestion des valeurs aberantts

```python
#iqr = Q3 - Q1
# outliers = Q1 - 1.5 * iqr ou Q3 + 1.5 * iqr
def finding_outliers(variable_name, data): 
    iqr = data[variable_name].quantile(0.75) - data[variable_name].quantile(0.25)
    lower = data[variable_name].quantile(0.25) - 1.5 * iqr
    upper = data[variable_name].quantile(0.75) + 1.5 * iqr
    return data[(data[variable_name] < lower) | (data[variable_name] > upper)]
```

#### Age

```python
finding_outliers('age', df).sort_values('age')
```

```python
iqr_age = df['age'].quantile(0.75) - df['age'].quantile(0.25)
iqr_age
```

```python
df.loc[(finding_outliers('age', df).index, 'age')] = df['age'].quantile(0.75) + 1.5 * iqr_age
```

```python
plt.figure()
sns.boxplot(y = 'age', data = df)
```

#### Fare

```python
finding_outliers('fare', df).sort_values('fare')
```

```python
iqr_age = df['fare'].quantile(0.75) - df['fare'].quantile(0.25)
iqr_age
```

```python
df.loc[(finding_outliers('fare', df).index, 'fare')] = df['fare'].quantile(0.75) + 1.5 * iqr_age
```

```python
plt.figure()
sns.boxplot(y = 'fare', data = df)
```

## Gestion de déséquilibres de classe

```python
from sklearn.utils import resample

X2 = X_train
X2['survived'] = y_train.values
X2.head()
```

```python
minority = X2[X2['survived'] == '1']
majority = X2[X2['survived'] == '0']
# Attetion : 'survived' est de type object, ne pas oublier d'ecrire X2[X2['survived'] == '1'] au lieu de X2[X2['survived'] == 1]. (les '')
```

```python
print(minority.shape)
print(majority.shape)
```

### Methode sur echantillonnage (upsamples)

```python
minority_upsampled = resample(minority, replace=True, n_samples = majority.shape[0])
```

```python
print(minority_upsampled.shape) 
print(majority.shape)
```

```python
upsampled = pd.concat([majority, minority_upsampled])
```

```python
upsampled.shape
```

### Méthode de sous échantillonnage (downsample)

```python
majority_downsampled = resample(majority, replace=False, n_samples = minority.shape[0])
```

```python
print(majority_downsampled.shape)
print(minority.shape)
```

```python
downsampled = pd.concat([minority, majority_downsampled])
```

```python
downsampled.shape
```

# 3.Modélisation et évaluation

### Train_test_split (Utilisation des données préparer):

```python
#Pour la suite nous allons utiliser upsampled

X_train_up = upsampled.drop('survived', axis=1)
y_train_up = upsampled['survived']
```

```python
print(X_train_up.shape)
print(y_train_up.shape)
```

#### Procédur d'évaluation

```python
from sklearn.metrics import f1_score, confusion_matrix, classification_report
from sklearn.model_selection import learning_curve
```

```python
def evaluation(model):

    model.fit(X_train_up, y_train_up)
    y_pred = model.predict(X_test)
    
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    N, train_up_score, val_score = learning_curve(model, X_train_up, y_train_up, 
                                              cv = 4, scoring = 'f1_macro', train_sizes=np.linspace(0.1, 1, 10))
    plt.figure(figsize = (12, 8))
    plt.plot(N, train_up_score.mean(axis=1), label = 'train_up_score')
    plt.plot(N, val_score.mean(axis=1), label = 'val_score')
    plt.legend()
    plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
preprocessor = make_pipeline(PolynomialFeatures(2))#, SelectKBest(f_classif, k = 5))
```

```python
RandomForestClassifier = make_pipeline(preprocessor, RandomForestClassifier(random_state = seed))
AdaBoostClassifier = make_pipeline(preprocessor, AdaBoostClassifier(random_state = seed))
DecisionTreeClassifier = make_pipeline(preprocessor, DecisionTreeClassifier(random_state = seed))
KNeighborsClassifier = make_pipeline(preprocessor, StandardScaler(), KNeighborsClassifier())
SVC = make_pipeline(preprocessor, StandardScaler(), SVC(random_state = seed))
LogisticRegression = make_pipeline(preprocessor, LogisticRegression(random_state = seed, max_iter=500))
```

```python
dict_of_models = {'RandomForestClassifier':RandomForestClassifier,
                  'AdaBoostClassifier':AdaBoostClassifier,
                  'DecisionTreeClassifier':DecisionTreeClassifier, 
                  'KNeighborsClassifier':KNeighborsClassifier, 
                  'SVC':SVC, 
                  'LogisticRegression':LogisticRegression
                 }
```

```python
for name, model in dict_of_models.items():
    print(name)
    evaluation(model)
    
```

# 4.Optimisation

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
SVC_1 = make_pipeline(preprocessor, StandardScaler(), SVC(random_state = seed))
```

```python
hyp_params = {'svc__C': [0.1, 1, 10, 100]}
```

```python
final_model = GridSearchCV(SVC_1, hyp_params, scoring= 'recall', cv = 4)

final_model.fit(X_train_up, y_train_up)

y_pred = final_model.predict(X_test)
```

```python
print(final_model.best_params_)

print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))
```
